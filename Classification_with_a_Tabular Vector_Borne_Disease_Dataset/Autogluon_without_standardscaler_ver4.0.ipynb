```python
import pandas as pd
import numpy as np
```

```python
#直接读入标准化后的train.csv
train = pd.read_csv("data/train.csv")
```

```python
train.head()
```

```python
def trans_prognois(col):
    name_list = list(col.value_counts().index)
    target_list = []
    for each in col:
         target_list.append(name_list.index(each)) #返回指定值的索引
    return target_list

train['prognosis'] = trans_prognois(train.prognosis)
train.head()
```

```python
train.to_csv('train_automl.csv')
```

```python
from autogluon.tabular import TabularDataset, TabularPredictor
```

```python
#定义需要drop的列和label列
id_1,unname,label  = 'id','Unnamed: 0','prognosis'
```

```python
train_data = TabularDataset('train_automl.csv')
```

```python
train_data.drop(columns=[id_1,unname],inplace=True)
```

```python
train_data
```

```python
predictor = TabularPredictor(label=label).fit(
            train_data)
```

```python
predictor.leaderboard(train_data, silent=True)
```

```python
#与对应类别，即labels进行一一对应
train_data_forl = pd.read_csv("train_standscaler.csv")
preds.columns  = train_data_forl.prognosis.value_counts().index
preds
```

```python
top3_cols = preds.apply(lambda x: ' '.join(x.nlargest(3).index.tolist()), axis=1)
top3_cols
```

```python
t = pd.read_csv('data/test.csv')
pd.DataFrame({'id':t.id,'prognosis':top3_cols}).set_index('id').to_csv('Autogluon_ver4.0.csv')
```
