```python
import pandas as pd
import numpy as np
```

```python
train = pd.read_csv("data/train.csv")
train.head()
```

```python
X = train.iloc[:,1:-1]
y = train.prognosis
```

```python
#删除id列和prognosis列
X.shape
```

```python
# #标准化数据
# from sklearn.preprocessing import StandardScaler

# sca = StandardScaler()
# X.iloc[:,:] = sca.fit_transform(X)
# X
```

```python
#pca降维
from sklearn.decomposition import PCA

pca = PCA(n_components=43)
new_X = pca.fit_transform(X)
new_X.shape
```

```python
#45个特征学习率到90%
np.cumsum(pca.explained_variance_ratio_)
```

```python
new_X = pd.DataFrame(new_X)
new_X
```

- 开始预测

```python
def trans_prognois(col):
    name_list = list(col.value_counts().index)
    target_list = []
    for each in col:
         target_list.append(name_list.index(each)) #返回指定值的索引
    return target_list

y_new = trans_prognois(y)
y_new
```

```python
from sklearn.svm import SVC

svc = SVC(kernel='rbf',probability=True,random_state=318)

#svc.fit(new_X,y_new)
```

```python
from sklearn.model_selection import cross_val_score

cross_val_score(svc,new_X,y_new).mean()
```

```python
svc.fit(new_X, y_new)
```

- 读test数据

```python
test = pd.read_csv('data/test.csv')
test
```

```python
X_test = pca.transform(test.iloc[:,1:])
pd.DataFrame(X_test)
```

```python
proba = svc.predict_proba(X_test)
proba
```

```python
output_df = pd.DataFrame(proba,columns = list(train.prognosis.value_counts().index))
output_df
```

```python
top3_cols = output_df.apply(lambda x: ' '.join(x.nlargest(3).index.tolist()), axis=1)
top3_cols
```

```python
t = pd.read_csv('data/test.csv')
pd.DataFrame({'id':t.id,'prognosis':top3_cols}).set_index('id').to_csv('pca_svc_ver3.0.csv')
```
