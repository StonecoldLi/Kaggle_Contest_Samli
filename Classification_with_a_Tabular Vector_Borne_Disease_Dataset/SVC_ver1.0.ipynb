### 对标签值做对应关系

```python
train_data.prognosis.value_counts()
```

```python
def trans_prognois(col):
    name_list = list(col.value_counts().index)
    target_list = []
    for each in col:
         target_list.append(name_list.index(each)) #返回指定值的索引
    return target_list
```

```python
#完成对y(标签列)的转换
y = trans_prognois(train_data.prognosis)
X = train_data.iloc[:,:-1]
```

```python
from sklearn.preprocessing import StandardScaler

def preprocess_data(df,train=True):
    #drop id column
    df.drop(columns=['id'], inplace=True)
    
    #standardScaler
    sc = StandardScaler()
    if train==True:
        df.iloc[:,:] = sc.fit_transform(df.iloc[:,:])
    
    return df
```

```python
X_train = preprocess_data(X)
```

```python
X_train
```

```python
from sklearn.svm import SVC

svc = SVC(kernel='rbf',probability=True,random_state=318)
X_new = svc.fit(X_train, y)
```

```python
from sklearn.model_selection import cross_val_score
print(cross_val_score(svc, X_train, y).mean())
```

- 查看每组数据在svc完成后预测为某一值的概率

```python
pd.DataFrame(svc.predict_proba(X_train),columns = list(train_data.prognosis.value_counts().index))
```

- 读取test_data

```python
test_data = pd.read_csv('data/test.csv')
```

```python
test_data.columns
```

```python
X_test = preprocess_data(test_data, train=False)
X_new = pd.DataFrame(X_test,columns=test_data.columns)
```

```python
output_df = pd.DataFrame(svc.predict_proba(X_new),columns = list(train_data.prognosis.value_counts().index))
```

```python
#选择预测概率为前三的作为output
def find_top3(row):
    top3 = row.nlargest(3)
    return top3
```

```python
top3_cols = output_df.apply(lambda x: ' '.join(x.nlargest(3).index.tolist()), axis=1)
top3_cols
```
