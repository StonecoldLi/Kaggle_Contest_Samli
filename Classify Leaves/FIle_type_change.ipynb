```python
import pandas as pd
```

```python
data = pd.read_csv('submission_densenet.csv')
data.head()
```

```python
data.columns
```

```python
data2 = pd.read_csv('submission_vgg11.csv',usecols=data.columns)
data2.head()
```

```python
data2['id'] = data2.id.astype('str')
```

```python
data2.dtypes
```

```python
def change_id(x):
    return "images/"+x+".jpg"
```

```python
data2["id"] = data2["id"].apply(change_id)
```

```python
data2
```

```python
top_cols = data2.iloc[:,1:].apply(lambda x: ' '.join(x.nlargest(1).index.tolist()), axis=1)
top_cols
```

```python
#data.to_csv("submission_densenet.csv",index=False)
pd.DataFrame({'image':data2.id,'label':top_cols}).set_index('image').to_csv('sub_ver1_vgg11_32.csv')
```
