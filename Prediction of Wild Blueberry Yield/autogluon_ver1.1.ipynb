```python
import pandas as pd
import numpy as np
```

- 处理train文件

```python
train = pd.read_csv("data/train.csv")
```

```python
X = train.iloc[:,:-1]
y = train['yield']
```

```python
X
```

```python
y
```

```python
X.drop(columns=['id'],inplace=True)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_new = sc.fit_transform(X)
X_new = pd.DataFrame(X_new,columns=X.columns)
X_new.shape
```

```python
train_new = pd.concat([X_new,y], axis=1)
train_new
```

```python
train_new.to_csv("autogluon_standardscaler_train.csv",index=False)
```

- autogluon处理

```python
from autogluon.tabular import TabularDataset, TabularPredictor
```

```python
train_temp = pd.read_csv("autogluon_standardscaler_train.csv")
train_temp.head()
```

```python
#规定train_data
train_data = TabularDataset("autogluon_standardscaler_train.csv")
```

```python
train_data
```

```python
#规定duplicate_columns, label
label = 'yield'
```

```python
predictor = TabularPredictor(label=label,problem_type='multiclass').fit(
            train_data)
```

```python
predictor.evaluate(train_data)
```

```python
#分类器排序
predictor.leaderboard(train_data, silent=True)
```

- 预测test数据

```python
test_data = TabularDataset('autogluon_standardscaler_test.csv')
```

```python
test_data
```

```python
preds = predictor.predict(test_data)
```

```python
preds
```

```python
pd.DataFrame({'id':test.id,'yield':preds}).set_index('id').to_csv('automl_multiclass_ver1.csv')
```
