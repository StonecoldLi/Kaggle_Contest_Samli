```python
import numpy as np
import pandas as pd
```

```python
train_data = pd.read_csv("train.csv")
test_data = pd.read_csv("test.csv")
```

```python
all_features = pd.concat((train_data.iloc[:, 1:-1], test_data.iloc[:, 1:]))
numeric_features = all_features.dtypes[all_features.dtypes != 'object'].index
all_features[numeric_features] = all_features[numeric_features].apply(
    lambda x: (x - x.mean()) / (x.std()))

# 在标准化数据之后，所有均值消失，因此我们可以将缺失值设置为0
all_features[numeric_features] = all_features[numeric_features].fillna(0)
all_features = pd.get_dummies(all_features, dummy_na=True)
all_features = all_features.astype('float32') #必须有这行！！！
all_features.shape

n_train = train_data.shape[0]
train_features = all_features[:n_train]
test_features = all_features[n_train:]
train_labels = train_data.SalePrice.values.reshape(-1, 1)
```

```python
from autogluon.tabular import TabularDataset, TabularPredictor
```

```python
train_price = pd.read_csv("train_automl.csv")
train_price.head()
```

```python
unname, label = 'Unnamed: 0','SalePrice'
```

```python
train_data = TabularDataset('train_automl.csv')
```

```python
predictor = TabularPredictor(label=label).fit(
            train_data.drop(columns=[unname]))
```

```python
test_data = TabularDataset('test_automl.csv')
```

```python
preds = predictor.predict(test_data.drop(columns=[unname]))
```

```python
test_data_raw = pd.read_csv('test.csv')
test_data_raw.head()
```

```python
pd.DataFrame({'id':test_data_raw.Id,'SalePrice':preds}).set_index('id').to_csv('sub_houseprice_automl.csv')
```
